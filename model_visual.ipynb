```python


# Load the model
model = load_model('cnnModel.keras')

# Assuming 'model' is your Keras model
plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=False, dpi=60, expand_nested=False)
```

```python
def count_layers(model):
    total_layers = 0
    for layer in model.layers:
        if isinstance(layer, tf.keras.Model):  # if the layer is a nested model
            total_layers += count_layers(layer)  # recursive call for nested models
        else:
            total_layers += 1
    return total_layers

print(f"Total number of layers (including nested): {count_layers(model)}")
```

```python
for i, layer in enumerate(model.layers):
    print(f"Layer {i + 1}: {layer.name}, Type: {layer.__class__.__name__}")
```

```python
total_layers = len(model.layers)
print(f"Total number of layers: {total_layers}")
```
